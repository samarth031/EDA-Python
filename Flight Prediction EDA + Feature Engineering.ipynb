```python
df = train.append(test)
print(df.tail())
```

```python
print(df.isnull().sum())
```

```python
df["Date"] = df["Date_of_Journey"].apply(lambda x: x.split("/")[0])
df["Month"] = df["Date_of_Journey"].apply(lambda x: x.split("/")[1])
df["Year"] = df["Date_of_Journey"].apply(lambda x: x.split("/")[2])
```

```python
df.head()
```

```python
cols = ["Date", "Month", "Year"]
for each_col in cols:
    df[each_col] = df[each_col].astype(int)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df.info()
```

```python
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
df["Arrival_Hour"] = df["Arrival_Time"].apply(lambda x: x.split(":")[0])
df["Arrival_Minute"] = df["Arrival_Time"].apply(lambda x: x.split(":")[1])
```

```python
df.head()
```

```python
cols = ["Arrival_Hour", "Arrival_Minute"]
for each_col in cols:
    df[each_col] = df[each_col].astype(int)
```

```python
df.info()
```

```python
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df["Dep_Time"] = df["Dep_Time"].apply(lambda x: x.split(" ")[0])
df["Dep_Hour"] = df["Dep_Time"].apply(lambda x: x.split(":")[0])
df["Dep_Minute"] = df["Dep_Time"].apply(lambda x: x.split(":")[1])
```

```python
df.drop("Dep_Time", axis=1, inplace=True)
```

```python
cols = ["Dep_Hour", "Dep_Minute"]
for each_col in cols:
    df[each_col] = df[each_col].astype(int)
```

```python
df.info()
```

```python
df["Total_Stops"].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4, "nan": 0})
```

```python
df.head()
```

```python
df.drop("Route", axis=1, inplace=True) # use Total_Stops instead of Route
```

```python
def duration_to_mins(x):
    x = x.replace("h", "").replace("m", "")
    x_hour = x.split(" ")[0]
    return x_hour
```

```python
df['Duration'] = df["Duration"].apply(lambda x: duration_to_mins(x))
```

```python
df.head()
```

```python
df["Airline"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Airline"] = le.fit_transform(df["Airline"])
df["Source"] = le.fit_transform(df["Source"])
df["Destination"] = le.fit_transform(df["Destination"])
df["Additional_Info"] = le.fit_transform(df["Additional_Info"])
```
